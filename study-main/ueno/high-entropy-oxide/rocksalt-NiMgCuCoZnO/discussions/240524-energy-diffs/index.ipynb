```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# テキストファイルからデータを読み込む
data = pd.read_csv('energy.txt', delim_whitespace=True, header=None, names=['Pre_Optimization_Energy', 'Post_Optimization_Energy'])

# データの相関係数を計算する
correlation = data.corr()

# 相関関係の結果を表示
print("Correlation coefficient between Pre-Optimization and Post-Optimization Energies:")
print(correlation)

# 散布図をプロットして視覚化
plt.scatter(data['Pre_Optimization_Energy'], data['Post_Optimization_Energy'])
plt.title('Scatter Plot of Pre-Optimization vs Post-Optimization Energies')
plt.xlabel('Energy (NSW=0) / eV')
plt.ylabel('Energy (NSW=2000) / eV')
plt.grid(True)
plt.show()
```
